```python
import pandas as pd
import os
from tabulate import tabulate
```

I have run the backest and saved the results in a csv files.<br>Now, I am just laoding the result csv files

```python
directory = '/Algorithmic-Trading/data/'
result_files_names = os.listdir(directory)

result_dfs = []
result_dic = {}
```

```python
for file in result_files_names:
    df = pd.read_csv(directory+file)
    # Use percentage to compare PnL
    df['pnl_percent'] = df.pnl / df.open_price
    df['dollar_invested'] =  (df['pnl_percent'] + 1).cumprod()
    result_dic[file.replace('.csv', '')] = {
    'name':file,
    'pnl':df.pnl.sum(),
    'pnl_percent_mean': df.pnl_percent.mean(),
    'return_on_$1':df['dollar_invested'].iloc[-1],
    'wining_percent': len(df[df.pnl > 0]) / len(df),
    'Average % gain per trade': df[df['pnl_percent'] > 0]['pnl_percent'].mean(),
    'Average % loss per trade': df[df['pnl_percent'] < 0]['pnl_percent'].mean(),
    'max_pnl %': df.pnl_percent.max(),
    'min_pnl %': df.pnl_percent.min(),
    'pnl_percent_std':df['pnl_percent'].std(),
    'number_of_trades': len(df),
    'df':df
    }
```

```python
r = pd.DataFrame.from_dict(result_dic, orient='index').drop(['df', 'name'], axis='columns')

for col in r.columns:
    if '%' in col or 'percent' in col:
        r[col] = r[col].map(lambda x: str(round(x * 100, 2)) + '%')
        
```

```python
r.sort_values('return_on_$1', ascending=False)
```
